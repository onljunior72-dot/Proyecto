```python
# @title 🚀 **CELDA ÚNICA DEFINITIVA: XFCE + Selkies + Cloudflare (Auto-detecta Selkies)**
import os, time, subprocess, re, glob

def print_step(msg):
    print("\n" + "="*60)
    print(f"👉 {msg}")
    print("="*60)

# -----------------------------------------------------------------------------
# PASO 1: Limpieza y reparación
# -----------------------------------------------------------------------------
print_step("PASO 1: Limpiando procesos y reparando dpkg")
!pkill -9 Xvfb 2>/dev/null
!pkill -9 xfce4-session 2>/dev/null
!pkill -9 xfwm4 2>/dev/null
!pkill -9 xfdesktop 2>/dev/null
!pkill -9 xfce4-panel 2>/dev/null
!pkill -9 Thunar 2>/dev/null
!pkill -9 selkies 2>/dev/null
!pkill -9 cloudflared 2>/dev/null
!rm -f /tmp/.X1-lock 2>/dev/null

!dpkg --configure -a
!apt-get update -qq
print("✅ Listo.")

# -----------------------------------------------------------------------------
# PASO 2: Instalar dependencias y XFCE
# -----------------------------------------------------------------------------
print_step("PASO 2: Instalando XFCE")
!export DEBIAN_FRONTEND=noninteractive
!apt-get install -y -qq xvfb xfce4 xfce4-goodies xfce4-terminal dbus-x11 wget curl net-tools
print("✅ XFCE instalado.")

# -----------------------------------------------------------------------------
# PASO 3: Descargar Selkies y Cloudflared
# -----------------------------------------------------------------------------
print_step("PASO 3: Obteniendo Selkies y Cloudflared")

# Descargar Selkies si no existe ningún directorio que contenga "selkies"
selkies_dirs = glob.glob('/content/*selkies*')
if not selkies_dirs:
    print("⬇️ Descargando Selkies...")
    !wget -q --show-progress https://github.com/selkies-project/selkies/releases/download/v1.6.2/selkies-gstreamer-portable-v1.6.2_amd64.tar.gz -O /content/selkies.tar.gz
    !tar -xzf /content/selkies.tar.gz -C /content/
    !rm /content/selkies.tar.gz
    # Volver a buscar la carpeta generada
    selkies_dirs = glob.glob('/content/*selkies*')
    if selkies_dirs:
        print(f"✅ Selkies extraído en: {selkies_dirs[0]}")
    else:
        print("❌ No se encontró la carpeta de Selkies después de extraer.")
else:
    print(f"✅ Selkies ya está en: {selkies_dirs[0]}")

# Cloudflared
if not os.path.exists('/content/cloudflared'):
    print("⬇️ Descargando cloudflared...")
    !wget -q --show-progress https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -O /content/cloudflared
    !chmod +x /content/cloudflared
    print("✅ cloudflared descargado.")
else:
    print("✅ cloudflared ya está presente.")

# -----------------------------------------------------------------------------
# PASO 4: Iniciar Xvfb y XFCE
# -----------------------------------------------------------------------------
print_step("PASO 4: Iniciando Xvfb y XFCE")
!pkill -9 Xvfb 2>/dev/null
!rm -f /tmp/.X1-lock 2>/dev/null

subprocess.Popen(['Xvfb', ':1', '-screen', '0', '1920x1080x24', '-ac', '+extension', 'GLX', '+render', '-noreset'],
                 stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
time.sleep(3)
os.environ['DISPLAY'] = ':1'
!service dbus start > /dev/null 2>&1
subprocess.Popen(['xfce4-session'], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
time.sleep(5)
print("✅ XFCE ejecutándose en DISPLAY=:1")

# -----------------------------------------------------------------------------
# PASO 5: Iniciar Selkies
# -----------------------------------------------------------------------------
print_step("PASO 5: Iniciando Selkies en puerto 8081")
!pkill -9 selkies 2>/dev/null

# Detectar la carpeta de Selkies
selkies_dirs = glob.glob('/content/*selkies*')
if not selkies_dirs:
    raise Exception("No se encontró la carpeta de Selkies. Ejecuta de nuevo la celda.")
selkies_dir = selkies_dirs[0]
print(f"📁 Usando carpeta Selkies: {selkies_dir}")

# Buscar el ejecutable (puede llamarse selkies-gstreamer-run o similar)
executable = None
for root, dirs, files in os.walk(selkies_dir):
    for f in files:
        if 'selkies-gstreamer-run' in f or f == 'selkies-gstreamer-run':
            executable = os.path.join(root, f)
            break
    if executable:
        break

if not executable:
    # Buscar cualquier archivo ejecutable que contenga "selkies"
    for root, dirs, files in os.walk(selkies_dir):
        for f in files:
            if 'selkies' in f and os.access(os.path.join(root, f), os.X_OK):
                executable = os.path.join(root, f)
                break
        if executable:
            break

if not executable:
    raise Exception(f"No se encontró el ejecutable de Selkies dentro de {selkies_dir}")

print(f"🚀 Ejecutable encontrado: {executable}")

# Lanzar Selkies
with open('/content/selkies.log', 'w') as log:
    subprocess.Popen([executable, '--port', '8081',
                      '--basic_auth_user', 'user', '--basic_auth_password', '123456',
                      '--enable_resize', 'false', '--enable_audio', 'true'],
                     stdout=log, stderr=subprocess.STDOUT, cwd=os.path.dirname(executable))
time.sleep(5)

# Verificar puerto
sock_check = !netstat -tlnp 2>/dev/null | grep 8081
if sock_check:
    print("✅ Selkies escuchando en puerto 8081")
else:
    print("⚠️ Selkies podría no haberse iniciado. Revisa /content/selkies.log")

# -----------------------------------------------------------------------------
# PASO 6: Cloudflare y URL
# -----------------------------------------------------------------------------
print_step("PASO 6: Iniciando túnel Cloudflare")
!pkill -9 cloudflared 2>/dev/null
!echo "" > /content/tunnel.log

subprocess.Popen(['/content/cloudflared', 'tunnel', '--url', 'http://localhost:8081'],
                 stdout=open('/content/tunnel.log', 'w'), stderr=subprocess.STDOUT)

print("⏳ Esperando enlace...")
url_encontrada = None
for i in range(30):
    time.sleep(1.5)
    try:
        with open('/content/tunnel.log', 'r') as f:
            log = f.read()
            match = re.search(r"https://[-a-zA-Z0-9.]*\.trycloudflare\.com", log)
            if match:
                url_encontrada = match.group(0)
                break
    except:
        pass

if url_encontrada:
    print("\n" + "🎉"*30)
    print("🎉 ¡MÁQUINA VIRTUAL LISTA!")
    print("🎉"*30)
    print(f"\n🌐 ENLACE: {url_encontrada}")
    print(f"👤 USUARIO: user")
    print(f"🔑 CONTRASEÑA: 123456")
else:
    print("\n⚠️ No se pudo obtener la URL. Revisa /content/tunnel.log")

print_step("✅ PROCESO COMPLETADO")
```

```python
# @title 🔄 **Keep-Alive con Python (bucle infinito)**
import time
from IPython.display import clear_output

print("⏳ Keep‑Alive activado. Presiona el botón ■ 'Interrumpir ejecución' para detenerlo.")
i = 0
while True:
    i += 1
    clear_output(wait=True)
    print(f"🔄 Keep‑Alive: sesión activa [{i}] - {time.strftime('%H:%M:%S')}")
    time.sleep(60)  # imprime cada 60 segundos
```
