```python
# @title 🚀 **CELDA ÚNICA DEFINITIVA CON PERSISTENCIA Y BACKUP**
import os, time, subprocess, re, glob, shutil
from datetime import datetime
import pickle

# ============================================
# CONFIGURACIÓN DE PERSISTENCIA
# ============================================
WORK_DIR = '/content/workspace'  # Directorio de trabajo persistente
BACKUP_DIR = '/content/drive/MyDrive/colab_backups'  # Backup en Google Drive
SESSION_FILE = f'{WORK_DIR}/session_state.pkl'  # Archivo de estado

def print_step(msg):
    print("\n" + "="*60)
    print(f"👉 {msg}")
    print("="*60)

# ============================================
# FUNCIONES DE BACKUP Y PERSISTENCIA
# ============================================

def setup_persistence():
    """Configura los directorios necesarios para persistencia"""
    # Montar Google Drive
    from google.colab import drive
    drive.mount('/content/drive', force_remount=True)

    # Crear directorios necesarios
    os.makedirs(WORK_DIR, exist_ok=True)
    os.makedirs(BACKUP_DIR, exist_ok=True)
    os.makedirs(f'{WORK_DIR}/downloads', exist_ok=True)
    os.makedirs(f'{WORK_DIR}/config', exist_ok=True)

    print(f"✅ Directorio de trabajo: {WORK_DIR}")
    print(f"✅ Backup en: {BACKUP_DIR}")
    return True

def save_session_state():
    """Guarda el estado actual de la sesión"""
    try:
        state = {
            'timestamp': datetime.now().isoformat(),
            'installed_packages': [],  # Puedes agregar paquetes instalados
            'selkies_version': None,
            'custom_files': []
        }

        # Guardar estado
        with open(SESSION_FILE, 'wb') as f:
            pickle.dump(state, f)

        print(f"✅ Estado guardado: {SESSION_FILE}")
        return True
    except Exception as e:
        print(f"⚠️ Error guardando estado: {e}")
        return False

def load_session_state():
    """Carga el estado previo de la sesión"""
    if os.path.exists(SESSION_FILE):
        try:
            with open(SESSION_FILE, 'rb') as f:
                state = pickle.load(f)
            print(f"✅ Estado cargado del {state['timestamp']}")
            return state
        except:
            print("⚠️ No se pudo cargar el estado previo")
    return None

def create_full_backup():
    """Crea un backup completo de todo el workspace"""
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    backup_name = f'colab_backup_{timestamp}'
    backup_path = f'{BACKUP_DIR}/{backup_name}'

    print(f"📦 Creando backup en: {backup_path}")

    try:
        # Crear backup comprimido
        shutil.make_archive(backup_path, 'zip', WORK_DIR)

        # También guardar logs importantes
        logs_backup = f'{BACKUP_DIR}/logs_{timestamp}.txt'
        with open(logs_backup, 'w') as f:
            f.write(f"Backup creado: {timestamp}\n")
            if os.path.exists('/content/selkies.log'):
                f.write("\n=== SELKIES LOG ===\n")
                f.write(open('/content/selkies.log').read())
            if os.path.exists('/content/tunnel.log'):
                f.write("\n=== TUNNEL LOG ===\n")
                f.write(open('/content/tunnel.log').read())

        print(f"✅ Backup completado: {backup_path}.zip")

        # Limpiar backups antiguos (mantener últimos 5)
        cleanup_old_backups()

        return backup_path
    except Exception as e:
        print(f"❌ Error creando backup: {e}")
        return None

def cleanup_old_backups(keep_last=5):
    """Mantiene solo los últimos N backups"""
    try:
        backups = sorted(glob.glob(f'{BACKUP_DIR}/colab_backup_*.zip'))
        if len(backups) > keep_last:
            for old_backup in backups[:-keep_last]:
                os.remove(old_backup)
                print(f"🗑️ Backup antiguo eliminado: {os.path.basename(old_backup)}")
        print(f"💾 Manteniendo últimos {keep_last} backups")
    except Exception as e:
        print(f"⚠️ Error limpiando backups: {e}")

def auto_backup_on_exit():
    """Función que se ejecuta automáticamente al detener la celda"""
    print("\n" + "🔴"*30)
    print("🔴 DETENIENDO CELDA - CREANDO BACKUP AUTOMÁTICO")
    print("🔴"*30)

    # Detener servicios
    print("🛑 Deteniendo servicios...")
    os.system('pkill -9 Xvfb xfce4-session selkies cloudflared 2>/dev/null')

    # Crear backup
    backup_path = create_full_backup()

    if backup_path:
        print(f"✅ Backup guardado en Google Drive: {backup_path}")
    else:
        print("❌ No se pudo crear el backup")

    save_session_state()
    print("✅ Proceso de backup completado")

# ============================================
# FUNCIÓN PARA RESTAURAR BACKUP
# ============================================

def restore_backup():
    """Restaura un backup previo desde Google Drive"""
    print_step("RESTAURANDO BACKUP")

    # Listar backups disponibles
    backups = glob.glob(f'{BACKUP_DIR}/colab_backup_*.zip')
    if not backups:
        print("❌ No se encontraron backups en Google Drive")
        return False

    print("\n📋 Backups disponibles:")
    for i, backup in enumerate(backups):
        backup_name = os.path.basename(backup)
        backup_size = os.path.getsize(backup) / (1024*1024)  # MB
        print(f"  {i+1}. {backup_name} ({backup_size:.2f} MB)")

    try:
        choice = int(input("\nSelecciona el número del backup a restaurar (0 para saltar): "))
        if choice == 0:
            return False

        selected_backup = backups[choice-1]
        print(f"📂 Restaurando: {selected_backup}")

        # Extraer backup
        shutil.unpack_archive(selected_backup, WORK_DIR)
        print(f"✅ Backup restaurado en {WORK_DIR}")

        # Restaurar estado
        load_session_state()
        return True
    except Exception as e:
        print(f"❌ Error restaurando: {e}")
        return False

# ============================================
# FUNCIONES PARA MANEJAR DESCARGAS
# ============================================

def download_with_persistence(url, filename=None):
    """Descarga un archivo y lo guarda en el workspace persistente"""
    if not filename:
        filename = url.split('/')[-1]

    download_path = f'{WORK_DIR}/downloads/{filename}'

    if os.path.exists(download_path):
        print(f"📁 Archivo ya existe: {download_path}")
        return download_path

    print(f"⬇️ Descargando: {filename}")
    os.system(f'wget -q --show-progress "{url}" -O "{download_path}"')

    if os.path.exists(download_path):
        print(f"✅ Descargado en: {download_path}")
        return download_path
    else:
        print(f"❌ Error descargando {url}")
        return None

def save_custom_file(content, filename, subdir=''):
    """Guarda un archivo personalizado en el workspace persistente"""
    save_dir = f'{WORK_DIR}/{subdir}' if subdir else WORK_DIR
    os.makedirs(save_dir, exist_ok=True)

    filepath = f'{save_dir}/{filename}'
    with open(filepath, 'w') as f:
        f.write(content)

    print(f"✅ Archivo guardado: {filepath}")
    return filepath

# ============================================
# SCRIPT PRINCIPAL MODIFICADO
# ============================================

def main():
    # Configurar persistencia
    setup_persistence()

    # Preguntar si restaurar backup
    print_step("¿RESTAURAR BACKUP PREVIO?")
    restore = input("¿Restaurar backup anterior? (s/N): ").lower() == 's'
    if restore:
        restore_backup()

    # PASO 1
    print_step("PASO 1: Limpiando procesos y reparando dpkg")
    os.system('pkill -9 Xvfb xfce4-session xfwm4 xfdesktop xfce4-panel Thunar selkies cloudflared 2>/dev/null')
    os.system('rm -f /tmp/.X1-lock 2>/dev/null')
    os.system('dpkg --configure -a')
    os.system('apt-get update -qq')
    print("✅ Listo.")

    # PASO 2 - Verificar si ya están instalados
    print_step("PASO 2: Instalando XFCE (si no está)")
    if not os.path.exists('/usr/bin/xfce4-session'):
        os.system('export DEBIAN_FRONTEND=noninteractive')
        os.system('apt-get install -y -qq xvfb xfce4 xfce4-goodies xfce4-terminal dbus-x11 wget curl net-tools')
    else:
        print("✅ XFCE ya está instalado")

    # PASO 3 - Usar workspace persistente
    print_step("PASO 3: Obteniendo Selkies y Cloudflared")

    # Mover selkies al workspace si existe en backup
    selkies_dirs = glob.glob(f'{WORK_DIR}/*selkies*')
    if not selkies_dirs:
        print("⬇️ Descargando Selkies...")
        os.system('wget -q --show-progress https://github.com/selkies-project/selkies/releases/download/v1.6.2/selkies-gstreamer-portable-v1.6.2_amd64.tar.gz -O /content/selkies.tar.gz')
        os.system('tar -xzf /content/selkies.tar.gz -C /content/')
        os.system('rm /content/selkies.tar.gz')

        # Mover al workspace persistente
        selkies_dirs = glob.glob('/content/*selkies*')
        if selkies_dirs:
            shutil.move(selkies_dirs[0], f'{WORK_DIR}/')
            print(f"✅ Selkies movido a {WORK_DIR}")
    else:
        print(f"✅ Selkies encontrado en: {selkies_dirs[0]}")

    # Cloudflared persistente
    cloudflared_path = f'{WORK_DIR}/cloudflared'
    if not os.path.exists(cloudflared_path):
        print("⬇️ Descargando cloudflared...")
        os.system(f'wget -q --show-progress https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -O {cloudflared_path}')
        os.system(f'chmod +x {cloudflared_path}')
        print("✅ cloudflared descargado.")
    else:
        print("✅ cloudflared ya está presente.")

    # PASO 4
    print_step("PASO 4: Iniciando Xvfb y XFCE")
    os.system('pkill -9 Xvfb 2>/dev/null')
    os.system('rm -f /tmp/.X1-lock 2>/dev/null')
    subprocess.Popen(['Xvfb', ':1', '-screen', '0', '1920x1080x24', '-ac', '+extension', 'GLX', '+render', '-noreset'],
                     stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    time.sleep(3)
    os.environ['DISPLAY'] = ':1'
    os.system('service dbus start > /dev/null 2>&1')
    subprocess.Popen(['xfce4-session'], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    time.sleep(5)
    print("✅ XFCE ejecutándose en DISPLAY=:1")

    # PASO 5
    print_step("PASO 5: Iniciando Selkies en puerto 8081")
    os.system('pkill -9 selkies 2>/dev/null')
    selkies_dir = glob.glob(f'{WORK_DIR}/*selkies*')[0]

    executable = None
    for root, dirs, files in os.walk(selkies_dir):
        for f in files:
            if 'selkies-gstreamer-run' in f or f == 'selkies-gstreamer-run':
                executable = os.path.join(root, f)
                break
        if executable:
            break

    with open(f'{WORK_DIR}/selkies.log', 'w') as log:
        subprocess.Popen([executable, '--port', '8081',
                          '--basic_auth_user', 'user', '--basic_auth_password', '123456',
                          '--enable_resize', 'false'],
                         stdout=log, stderr=subprocess.STDOUT, cwd=os.path.dirname(executable))
    time.sleep(5)

    # PASO 6
    print_step("PASO 6: Iniciando túnel Cloudflare")
    os.system('pkill -9 cloudflared 2>/dev/null')
    with open(f'{WORK_DIR}/tunnel.log', 'w') as f:
        subprocess.Popen([cloudflared_path, 'tunnel', '--url', 'http://localhost:8081'],
                        stdout=f, stderr=subprocess.STDOUT)

    print("⏳ Esperando enlace...")
    url_encontrada = None
    for i in range(30):
        time.sleep(1.5)
        try:
            with open(f'{WORK_DIR}/tunnel.log', 'r') as f:
                log = f.read()
                match = re.search(r"https://[-a-zA-Z0-9.]*\.trycloudflare\.com", log)
                if match:
                    url_encontrada = match.group(0)
                    break
        except:
            pass

    if url_encontrada:
        print("\n" + "🎉"*30)
        print("🎉 ¡MÁQUINA VIRTUAL LISTA!")
        print("🎉"*30)
        print(f"\n🌐 ENLACE: {url_encontrada}")
        print(f"👤 USUARIO: user")
        print(f"🔑 CONTRASEÑA: 123456")
        print(f"💾 WORKSPACE: {WORK_DIR}")

        # Guardar URL en el workspace
        with open(f'{WORK_DIR}/current_url.txt', 'w') as f:
            f.write(url_encontrada)

        # Guardar estado
        save_session_state()
    else:
        print("\n⚠️ No se pudo obtener la URL")

    print_step("✅ PROCESO COMPLETADO")

    # Configurar auto-backup al salir (para notebooks de Colab)
    import atexit
    atexit.register(auto_backup_on_exit)

    print("\n💡 **COMANDOS ÚTILES:**")
    print("  # Crear backup manual")
    print("  create_full_backup()")
    print("  # Listar archivos descargados")
    print("  !ls -la /content/workspace/downloads")
    print("  # Ver logs")
    print(f"  !cat {WORK_DIR}/selkies.log")
    print(f"  !cat {WORK_DIR}/tunnel.log")
    print("  # Subir archivos al workspace")
    print("  from google.colab import files")
    print("  uploaded = files.upload()")
    print("  # Mover a workspace persistente")
    print("  !mv uploaded_file.txt /content/workspace/downloads/")

# ============================================
# EJECUTAR SCRIPT PRINCIPAL
# ============================================

if __name__ == "__main__":
    main()
```
